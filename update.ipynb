```python
# Step 10: Detect and show outliers in pressure and leak_rate

import pandas as pd
import matplotlib.pyplot as plt

# Use the correct column names
pressure_col = "pressure (bar)"
leak_col = "leak_rate (cm3/mn)"

# Columns to analyze
cols_to_check = [pressure_col, leak_col]

outlier_rows = []

for col in cols_to_check:
    # Calculate Q1, Q3 and IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    # Outlier limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = data[(data[col] < lower_limit) | (data[col] > upper_limit)].copy()

    # Add information
    outliers["outlier_column"] = col
    outliers["lower_limit"] = lower_limit
    outliers["upper_limit"] = upper_limit

    print("\nColumn:", col)
    print("Q1:", Q1)
    print("Q3:", Q3)
    print("IQR:", IQR)
    print("Lower limit:", lower_limit)
    print("Upper limit:", upper_limit)
    print("Number of outliers:", len(outliers))

    display(outliers)

    outlier_rows.append(outliers)

# Combine all outliers in one table
all_outliers = pd.concat(outlier_rows, ignore_index=True)

print("\nAll detected outliers:")
display(all_outliers)
```

```python
# Cleaning Step 7: Treat empty flow_leak_pa_s using real conversion factor from the dataset

# 1. Select valid rows where leak_rate and flow_leak_pa_s are available
valid_rows = clean_data[
    clean_data["leak_rate"].notnull() &
    clean_data["flow_leak_pa_s"].notnull() &
    (clean_data["leak_rate"] > 0)
].copy()

# 2. Calculate the conversion ratio for each valid row
valid_rows["conversion_ratio"] = valid_rows["flow_leak_pa_s"] / valid_rows["leak_rate"]

# 3. Use the median ratio as conversion factor
conversion_factor = valid_rows["conversion_ratio"].median()

print("Number of valid rows used:", len(valid_rows))
print("Calculated conversion factor:", conversion_factor)

# 4. Check missing values before cleaning
print("\nEmpty flow_leak_pa_s before cleaning:")
print(clean_data["flow_leak_pa_s"].isnull().sum())

# 5. Replace empty flow_leak_pa_s using leak_rate × conversion_factor
clean_data.loc[
    clean_data["flow_leak_pa_s"].isnull(),
    "flow_leak_pa_s"
] = clean_data["leak_rate"] * conversion_factor

# 6. Check missing values after cleaning
print("\nEmpty flow_leak_pa_s after cleaning:")
print(clean_data["flow_leak_pa_s"].isnull().sum())

# 7. Verify alarm cases
alarm_clean = (
    clean_data["alarm"]
    .fillna("NO_ALARM")
    .astype(str)
    .str.strip()
    .str.upper()
)

print("\nLARGE LEAK TEST cases after cleaning:")
display(clean_data[alarm_clean == "LARGE LEAK TEST"][[
    "leak_rate", "flow_leak_pa_s", "alarm", "result", "quality_class"
]])

print("\nPRESSURE LOW cases after cleaning:")
display(clean_data[alarm_clean == "PRESSURE LOW"][[
    "leak_rate", "flow_leak_pa_s", "alarm", "result", "quality_class"
]])
```

```python
# Training Step 1: Separate input features X and target y
# Version 2: Remove reference, channel, date and time

# Target column
target_col = "quality_class"

# Columns not used as model inputs
columns_to_drop = [
    target_col,
    "reference",
    "channel",
    "date",
    "time"
]

# Create input data X
X = clean_data.drop(columns=columns_to_drop)

# Create target data y
y = clean_data[target_col]

# Show results
print("Input features X:")
print(X.columns.tolist())

print("\nTarget y:")
print(y.name)

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)

print("\nTarget classes:")
print(y.value_counts())
```

```python
# Training Step 2: Split the dataset into training and testing data

from sklearn.model_selection import train_test_split

# Split data: 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Show shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# Check class distribution in training data
print("\nTraining class distribution:")
print(y_train.value_counts())

# Check class distribution in test data
print("\nTest class distribution:")
print(y_test.value_counts())
```

```python
# Training Step 3: Encode target classes

from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Create label encoder
label_encoder = LabelEncoder()

# Fit on training target and transform both train and test target
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Show class mapping
class_mapping = pd.DataFrame({
    "Class Name": label_encoder.classes_,
    "Encoded Value": range(len(label_encoder.classes_))
})

print("Quality class encoding:")
display(class_mapping)

# Verify encoded values
print("First encoded y_train values:")
print(y_train_encoded[:10])
```

```python
# Training Step 4: Prepare preprocessing for input features X

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Detect numerical and categorical columns
numeric_features = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = X_train.select_dtypes(include=["object"]).columns.tolist()

print("Numerical features:")
print(numeric_features)

print("\nCategorical features:")
print(categorical_features)

# Preprocessing for numerical columns
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# Preprocessing for categorical columns
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

# Combine preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

print("\nPreprocessor is ready.")
```

Train random forest modele

```python
# Training Step 5: Train Random Forest model

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

rf_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        class_weight="balanced"
    ))
])

rf_model.fit(X_train, y_train_encoded)

y_pred_rf = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test_encoded, y_pred_rf)

print("Random Forest Accuracy:", rf_accuracy)

print("\nClassification Report - Random Forest:")
print(classification_report(
    y_test_encoded,
    y_pred_rf,
    target_names=label_encoder.classes_
))

cm_rf = confusion_matrix(y_test_encoded, y_pred_rf)

plt.figure(figsize=(7, 5))
sns.heatmap(
    cm_rf,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)

plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

```python
# Training Step 6: Train Improved SVM model with GridSearchCV

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

svm_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", SVC(
        kernel="rbf",
        class_weight="balanced",
        random_state=42
    ))
])

param_grid_svm = {
    "classifier__C": [0.1, 1, 10, 100],
    "classifier__gamma": ["scale", 0.01, 0.1, 1]
}

svm_grid = GridSearchCV(
    estimator=svm_pipeline,
    param_grid=param_grid_svm,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1
)

svm_grid.fit(X_train, y_train_encoded)

svm_model = svm_grid.best_estimator_

print("Best SVM parameters:")
print(svm_grid.best_params_)

y_pred_svm = svm_model.predict(X_test)

svm_accuracy = accuracy_score(y_test_encoded, y_pred_svm)

print("Improved SVM Accuracy:", svm_accuracy)

print("\nClassification Report - Improved SVM:")
print(classification_report(
    y_test_encoded,
    y_pred_svm,
    target_names=label_encoder.classes_
))

cm_svm = confusion_matrix(y_test_encoded, y_pred_svm)

plt.figure(figsize=(7, 5))
sns.heatmap(
    cm_svm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)

plt.title("Confusion Matrix - Improved SVM")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

```python
# Training Step 8: Train XGBoost model

from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create XGBoost pipeline
xgb_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", XGBClassifier(
        n_estimators=200,
        learning_rate=0.1,
        max_depth=4,
        random_state=42,
        eval_metric="mlogloss"
    ))
])

# Train the model
xgb_model.fit(X_train, y_train_encoded)

# Predict on test data
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate accuracy
xgb_accuracy = accuracy_score(y_test_encoded, y_pred_xgb)

print("XGBoost Accuracy:", xgb_accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(
    y_test_encoded,
    y_pred_xgb,
    target_names=label_encoder.classes_
))

# Confusion matrix
cm_xgb = confusion_matrix(y_test_encoded, y_pred_xgb)

plt.figure(figsize=(7, 5))
sns.heatmap(
    cm_xgb,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)

plt.title("Confusion Matrix - XGBoost")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

```python
# Training Step 9: Train Logistic Regression model

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create Logistic Regression pipeline
logreg_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ))
])

# Train the model
logreg_model.fit(X_train, y_train_encoded)

# Predict on test data
y_pred_logreg = logreg_model.predict(X_test)

# Evaluate accuracy
logreg_accuracy = accuracy_score(y_test_encoded, y_pred_logreg)

print("Logistic Regression Accuracy:", logreg_accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(
    y_test_encoded,
    y_pred_logreg,
    target_names=label_encoder.classes_
))

# Confusion matrix
cm_logreg = confusion_matrix(y_test_encoded, y_pred_logreg)

plt.figure(figsize=(7, 5))
sns.heatmap(
    cm_logreg,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)

plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

```python
# Training Step 10: Feature Importance for the 4 models using Permutation Importance

from sklearn.inspection import permutation_importance
import pandas as pd
import matplotlib.pyplot as plt

models = {
    "Random Forest": rf_model,
    "SVM": svm_model,
    "XGBoost": xgb_model,
    "Logistic Regression": logreg_model
}

importance_results = {}

for model_name, model in models.items():

    print("=" * 70)
    print("Feature Importance for:", model_name)

    result = permutation_importance(
        model,
        X_test,
        y_test_encoded,
        n_repeats=10,
        random_state=42,
        scoring="accuracy"
    )

    importance_df = pd.DataFrame({
        "Feature": X_test.columns,
        "Importance": result.importances_mean,
        "Std": result.importances_std
    }).sort_values(by="Importance", ascending=False)

    importance_results[model_name] = importance_df

    display(importance_df.round(4))

    plt.figure(figsize=(8, 5))
    plt.barh(importance_df["Feature"], importance_df["Importance"])
    plt.gca().invert_yaxis()
    plt.title(f"Feature Importance - {model_name}")
    plt.xlabel("Decrease in Accuracy")
    plt.ylabel("Feature")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.show()
```

```python
# Compare Train Accuracy and Test Accuracy

from sklearn.metrics import accuracy_score
import pandas as pd

models = {
    "Random Forest": rf_model,
    "Improved SVM": svm_model,
    "XGBoost": xgb_model,
    "Logistic Regression": logreg_model
}

train_test_results = []

for model_name, model in models.items():

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_train_encoded, y_train_pred)
    test_acc = accuracy_score(y_test_encoded, y_test_pred)

    train_test_results.append({
        "Model": model_name,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc,
        "Difference": train_acc - test_acc
    })

train_test_results_df = pd.DataFrame(train_test_results)

display(train_test_results_df.round(4))
```
